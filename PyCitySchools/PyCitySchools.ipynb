PyCity Schools

Written report:
- The analysis consists in the merging of two data bases with information regarding 15 schools, focused on reading and maths scores. 

The conclusions I found from the analysis are the following:
- The average reading score is higher than the average math score which means that more students are proficient in reading than in math.
- The Cabrera High School has the highest scores in terms of both reading and math scores. 
- While Rodriguez Highscool has the lowest ratings.
- According to the data, less spending yields higher scores.
- The data also states that the bigger the school the lower the scores the students get.

```python
#setup
import pandas as pd
from pathlib import Path
```

```python
#load the databases
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# merge files by school name
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#create array of unique school names
school_unique_name = school_data_complete['school_name'].unique()

#variable to store the length of unique school names to count them
total_schools = len(school_unique_name)
total_schools
```

```python
#calculate the number of students
number_students = school_data_complete['student_name'].count()

number_students
```

```python
# Calculate the sum of the budget of all the schools, using the school data df
sum_of_budget = school_data['budget'].sum()
sum_of_budget
```

```python
#calculate the average math score
avg_math_score = school_data_complete['math_score'].mean()
avg_math_score
```

```python
#calculate the average reading score
avg_reading_score = school_data_complete['reading_score'].mean()
avg_reading_score
```

```python
#calculate the percentage of students tha passed math
percent_pass_math = school_data_complete.loc[school_data_complete['math_score']>= 70]['math_score'].count()
percentage_pass_math = percent_pass_math/number_students
percentage_pass_math
```

```python
#calculate the percentage of students tha passed reading
percent_pass_reading = school_data_complete.loc[school_data_complete['reading_score']>= 70]['reading_score'].count()
percentage_pass_reading = percent_pass_reading/number_students
percentage_pass_reading
```

```python
# calculate overall passing students
overall_passing = school_data_complete[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)]['student_name'].count()/number_students
overall_passing
```

```python
# display results into a dataframe and add the style format 
district_df = pd.DataFrame({"Total number of schools": [total_schools],
    "Total students": [number_students],
    "Total budget": [sum_of_budget],
    "Average math score": [avg_math_score],
    "Average reading score": [avg_reading_score],
    "% passing math":[percentage_pass_math],
    "% passing reading": [percentage_pass_reading],
    "Overall passing rate": [overall_passing]})

district_df.style.format({"Total budget": "${:,.2f}", 
                       "Average math score": "{:.1f}", 
                       "Average reading score": "{:.1f}", 
                       "% passing math": "{:.1%}", 
                       "% passing reading": "{:.1%}", 
                       "Overall passing rate": "{:.1%}"})
```

School Summary

```python
#creat groups by schools
groups_schools = school_data_complete.set_index('school_name').groupby(['school_name'])
groups_schools.head()
```

```python
# see the school types
school_types = school_data.set_index('school_name')['type']
school_types
```

```python
# get the total students per school
students_per_school = groups_schools['student_name'].count()
students_per_school
```

```python
# school budget
school_budget = school_data.set_index('school_name')['budget']
school_budget
```

```python
budget_per_student = school_budget / school_data.set_index('school_name')['size']
budget_per_student
```

```python
#average scores by school
average_math_school = groups_schools['math_score'].mean()
average_read_school = groups_schools['reading_score'].mean()
print(average_math_school)
print(average_read_school)
```

```python
# Percentage of passing scores

percentage_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['student_name'].count()/students_per_school 
percentage_reading = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['student_name'].count()/students_per_school 
percentage_overall = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('school_name')['student_name'].count()/students_per_school 
```

```python
#display the school dataframe

school_df= pd.DataFrame({"School type": school_types,
    "Total students": students_per_school,
    "Total school budget": school_budget,
    "Per student budget": budget_per_student,
    "Average math score": average_math_school,
    "Average reading score": average_read_school,
    'Percentage passing math': percentage_math,
    'Percentage passing reading': percentage_reading,
    "Overall passing percentage": percentage_overall})

school_df_f= school_df.style.format({'Total students': '{:,}', 
                          "Total school budget": "${:,}", 
                          "Per student budget": "${:.0f}",
                          'Average math score': "{:.1f}", 
                          'Average reading score': "{:.1f}", 
                          "Percentage passing math": "{:.1%}", 
                          "Percentage passing reading": "{:.1%}", 
                          "Overall passing percentage": "{:.1%}"})

school_df_f
```

Math Scores by Grade

```python
# create variabels for each grade per school and math scores
ninth_grade_math = school_data_complete.loc[school_data_complete['grade'] == '9th'].groupby('school_name')["math_score"].mean()
tenth_grade_math = school_data_complete.loc[school_data_complete['grade'] == '10th'].groupby('school_name')["math_score"].mean()
eleventh_grade_math = school_data_complete.loc[school_data_complete['grade'] == '11th'].groupby('school_name')["math_score"].mean()
twelfth_grade_math = school_data_complete.loc[school_data_complete['grade'] == '12th'].groupby('school_name')["math_score"].mean()

math_scores_by_grade = pd.DataFrame({ "9th": ninth_grade_math,
        "10th": tenth_grade_math,
        "11th": eleventh_grade_math,
        "12th": twelfth_grade_math})

math_scores_by_grade = math_scores_by_grade[['9th', '10th', '11th', '12th']]
math_scores_by_grade.index.name = "School"

#add format
math_scores_by_grade.style.format({'9th': '{:.1f}', 
                          "10th": '{:.1f}', 
                          "11th": "{:.1f}", 
                          "12th": "{:.1f}"})
```

Reading Scores by Grade

```python
# create variabels for each grade per school and reading scores
ninth_grade_reading = school_data_complete.loc[school_data_complete['grade'] == '9th'].groupby('school_name')["reading_score"].mean()
tenth_grade_reading = school_data_complete.loc[school_data_complete['grade'] == '10th'].groupby('school_name')["reading_score"].mean()
eleventh_grade_reading = school_data_complete.loc[school_data_complete['grade'] == '11th'].groupby('school_name')["reading_score"].mean()
twelfth_grade_reading = school_data_complete.loc[school_data_complete['grade'] == '12th'].groupby('school_name')["reading_score"].mean()

reading_scores_by_grade = pd.DataFrame({ "9th": ninth_grade_reading,
        "10th": tenth_grade_reading,
        "11th": eleventh_grade_reading,
        "12th": twelfth_grade_reading})

reading_scores_by_grade = reading_scores_by_grade[['9th', '10th', '11th', '12th']]
reading_scores_by_grade.index.name = "School"

#add format
reading_scores_by_grade.style.format({'9th': '{:.1f}', 
                          "10th": '{:.1f}', 
                          "11th": "{:.1f}", 
                          "12th": "{:.1f}"})
```

Scores by School Spending

```python
# create spending bins
spending_bins = [0, 585, 630, 645, 680]
labels = ['< $585', "$585 - 630", "$630 - 645", "$645- $680"]
school_data_complete['spendidg_bins'] = pd.cut(school_data_complete['budget']/school_data_complete['size'], spending_bins, labels = labels)

#group by spending
school_spending = school_data_complete.groupby('spending_bins')

# calculate means and percentages
spending_math_scores = school_spending['math_score'].mean()
spending_reading_scores = school_spending['reading_score'].mean()
spending_passing_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby('spending_bins')['student_name'].count()/school_spending['student_name'].count()
spending_passing_reading = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('spending_bins')['student_name'].count()/school_spending['student_name'].count()
overall_passing_spending = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('spending_bins')['student_name'].count()/school_spending['student_name'].count()

# create a dataframe
spending_summary = pd.DataFrame({"Average math score": spending_math_scores,
    "Average reading score": spending_reading_scores,
    '% passing math': spending_passing_math,
    '% passing reading': spending_passing_reading,
    "% overall passing": overall_passing_spending})

spending_summary.index.name = "Spending ranges"

#Apply format
spending_summary.style.format({'Average math score': '{:.1f}', 
                              'Average reading score': '{:.1f}', 
                              '% passing math': '{:.1%}', 
                              '% passing reading':'{:.1%}', 
                              '% overall passing': '{:.1%}'})

```

Scores by School Size

```python
# create size bins
size_bins = [0, 1000, 2000, 5000]
labels_size = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
school_data_complete['size_bins'] = pd.cut(school_data_complete['size'], size_bins, labels = labels_size)

#group by size
school_size = school_data_complete.groupby('size_bins')

# calculate means and percentages
size_math_scores = school_size['math_score'].mean()
size_reading_scores = school_size['reading_score'].mean()
size_passing_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby('size_bins')['student_name'].count()/school_size['student_name'].count()
size_passing_reading = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('size_bins')['student_name'].count()/school_size['student_name'].count()
overall_passing_size = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('size_bins')['student_name'].count()/school_size['student_name'].count()

# create a dataframe
size_summary = pd.DataFrame({"Average math score": size_math_scores,
    "Average reading score": size_reading_scores,
    '% passing math': size_passing_math,
    '% passing reading': size_passing_reading,
    "% overall passing": overall_passing_size})

size_summary.index.name = "Size ranges"

#Apply format
size_summary.style.format({'Average math score': '{:.1f}', 
                              'Average reading score': '{:.1f}', 
                              '% passing math': '{:.1%}', 
                              '% passing reading':'{:.1%}', 
                              '% overall passing': '{:.1%}'})
```
